```python
imagePath = "drive/MyDrive/Archive/data/image1_FC_34243.jpg"
imageData = image.imread(imagePath)
imageVisualization = plt.imshow(imageData)
plt.show()
```

```python
plt.figure(figsize=(20, 18))

fig, ax = plt.subplots()
ax.imshow(imageData)

JsonDataPathFile = open("drive/MyDrive/Archive/data/image1_FC_34243_judgements.json")
JasonData = json.load(JsonDataPathFile)

for i in JasonData['0']:
    
    min_X_axisData = 100000
    max_X_axisData = 0
    
    min_Y_axisData = 100000
    max_Y_axisData = 0
    
    
    for coords in i['boundaries'][0]['boundaryPoints']:
        if min_X_axisData > coords['coords'][0]:
            min_X_axisData = coords['coords'][0]
            
        if max_X_axisData < coords['coords'][0]:
            max_X_axisData = coords['coords'][0]
        
        if min_Y_axisData > coords['coords'][1]:
            min_Y_axisData = coords['coords'][1]
            
        if max_Y_axisData < coords['coords'][1]:
            max_Y_axisData = coords['coords'][1]
            
    drawRactangle = patches.Rectangle((min_X_axisData, min_Y_axisData), (max_X_axisData -min_X_axisData), (max_Y_axisData - min_Y_axisData), 
                                     linewidth = 1, edgecolor = 'r', facecolor= 'none')
    ax.add_patch(drawRactangle)
    print("max_X_axisData", max_X_axisData, "||", "min_X_axisData", min_X_axisData)
    print("max_Y_axisData", max_Y_axisData, "||", "min_Y_axisData", min_Y_axisData)
    
    
plt.show()
```
